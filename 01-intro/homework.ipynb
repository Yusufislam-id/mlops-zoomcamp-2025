```python
df = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet")
```

```python
df.head()
```

```python
num_columns = df.shape[1]
num_rows = df.shape[0]
print("Number of columns:", num_columns)
print("Number of rows:", num_rows)
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Calculate duration in minutes
df['duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
print(df[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'duration']].head())
```

```python
df.head()
```

```python
#standard deviation of the trips duration in January
std_dev = df['duration'].std()
print("Standard Deviation:", std_dev)
```

```python
# remove outliers
filtered_df = df[(df['duration'] >= 1) & (df['duration'] <= 60)]
print(filtered_df['duration'].describe())
```

```python
# Total original records
total_records = len(df)

# Records after filtering
filtered_records = len(filtered_df)

# Fraction remaining
fraction_remaining = filtered_records / total_records * 100

print(f"Fraction: {fraction_remaining:.0f}%")
```
